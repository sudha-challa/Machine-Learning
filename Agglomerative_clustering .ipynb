```python
#import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
```

```python
data_points=np.array([
    [0.40,0.53],
    [0.22,0.38],
    [0.35,0.32],
    [0.26,0.19],
    [0.08,0.41],
    [0.45,0.30]
])
labels=["p1", "p2", "p3", "p4", "p5", "p6",]
```

```python
# perform heirarchical clustering
linked=sch.linkage(data_points, method='single', metric='euclidean')
```

```python
plt.figure(figsize=(10,5))
sch.dendrogram(linked, labels=labels, orientation="top", distance_sort='ascending', show_leaf_counts=True)
plt.title("Agglomerative Clustering Dendrogram(single Linkage)")
plt.xlabel("Simple Points")
plt.ylabel("Distance")
# plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# perform heirarchical clustering
linked=sch.linkage(data_points, method='complete', metric='euclidean')
```

```python
plt.figure(figsize=(10,5))
sch.dendrogram(linked, labels=labels, orientation="top", distance_sort='ascending', show_leaf_counts=True)
plt.title("Agglomerative Clustering Dendrogram(complete Linkage)")
plt.xlabel("Simple Points")
plt.ylabel("Distance")
# plt.grid(True)
plt.tight_layout()
plt.show()
```
